```python
cel = [n for n in range(10, 500, 30)]
conver = map(lambda n: (n * 1.8)+32, cel)
fah = list(conver)
print(fah)

lista1 = [n for n in range(1, 5)]
lista2 = [n for n in range(6, 10)]
lista3 = map(lambda x, y: x + y, lista1, lista2)
final = list(lista3)
print(final)

radios = [r for r in range(2, 8)]
areas = map(lambda r: 3.1416 * r**2, radios)
resultado = list(areas)
print(resultado)

valor = {"pan": 1000, "leche": 2500, "café": 5000}
lista = [(nombre, precio) for nombre, precio in valor.items()]
lista_i = map(lambda n: (n[0], n[1] * 1.10), lista)
lista_f = list(lista_i)
print(lista_f)

mat = [
    [1,2],
    [3,4],
    [5,6]
]

mult_por_10 = [[n * 10 for n in fil] for fil in mat]
print(mult_por_10)
```
