```python
from functools import reduce

numeros = [12, 24, 36, 48]

resultado = reduce(lambda n, z: n * z, numeros)
print(resultado)

palabras = ["arroz", "puerta", "computador", "telefono"]
unir = reduce(lambda n, z: n + z, palabras)
print(palabras)

numeros = [2, 4, 12, 3, 1]

mayor = lambda x, y: x if x > y else y
mayor_f = reduce(mayor, numeros)
print(mayor_f)
```
