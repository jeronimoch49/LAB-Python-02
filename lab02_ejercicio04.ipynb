```python
palabras = ["carro", "laptop", "llave", "python", "gato"]

mayores4 = filter(lambda n: len(n) > 4, palabras)
mayores4 = list(mayores4)
print(mayores4)

datos = [10, None, "Python", None, True, 42]

acepta = filter(lambda n: n is not None, datos)
acepta = list(acepta)
print(acepta)

vocales = ["arbol", "perro", "uva", "mañana", "iglesia"]

acepta = list(filter(lambda n: n[0] == "a" or n[0] == "e" or n[0] == "i" or n[0] == "o" or n[0] == "u", vocales))
print(acepta)

palabras = ["oso", "raiz", "radar", "espejo", "ana"]

palin = list(filter(lambda n: n == n[::-1], palabras))
print(palin)

numeros = [15, 10, 89, 65, 37, 47]

terminan = list(filter(lambda n: n % 10 == 5, numeros))
print(terminan)

productos = [
    {"nombre": "Teclado", "precio": 80},
    {"nombre": "Mouse", "precio": 40},
    {"nombre": "Monitor", "precio": 300},
    {"nombre": "Webcam", "precio": 150}
]

mayor_100 = list(filter(lambda n: n["precio"] > 100, productos))
print(mayor_100)
```
