```python
numeros = [2, 5, 4, 9]

ordenar = sorted([n ** 2 for n in numeros if n % 2 == 0])
print(ordenar)

numeros = [67, 200, 101, 302, 42]

elevar = [str(n**3) for n in numeros if n**3 > 100]
ordenado = sorted(elevar, key=int)
print(ordenado)


numeros = [2, 12, 20, 16, 5]
ordenar = [n**2 for n in numeros if n > 10]
orden = sorted(ordenar, reverse=True)
print(orden)
```
