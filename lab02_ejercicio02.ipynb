```python
numeros = [1,2,3,4,5]

lista_tuplas = [(n, n**2) for n in numeros]
print(lista_tuplas)

cubos = [n**3 for n in range(1, 20) if n > 10]
print(cubos)

pares = [n for n in range(1,20) if n % 2 == 0]
print(pares)

numero = [n for n in range(10,50,5)]
dado = [f"valor: {n}" for n in numero]
print(dado)

cel = [n for n in range(2, 17, 3)]
far = [n * 9.5 +32 for n in cel]
print(far)

palabras = ["perro", "computadora", "python", "mar", "algoritmo", "ia", "programacion"]
rango = [palabra for palabra in palabras if len(palabra) > 5]
print(rango)

pa = ["mama", "consola", "llave", "programa", "celular"]
primera = [inicial[0] for inicial in pa]
print(primera)

lista = [n for n in range(1,10)]
par_o_im = ["par" if n % 2 == 0 else "impar" for n in lista]
print(par_o_im)

mat = [[1,2,3],[4,5,6],[7,8,9]]
aplanar = [elemento for fila in mat for elemento in fila]
print(aplanar)
```
