```python
palabras = ["abuela", "xilofono", "jirafa", "barco"]

print(sorted(palabras))

tupla = [(2, "carro"), (4, "moto"), (1, "bicicleta"), (3, "patineta")]

print(sorted(tupla))

productos = [
      ("Camisa", 30),
      ("Pantalón", 55),
      ("Medias", 10),
      ("Chaqueta", 80)
]

ordenar_por_precio = sorted(productos, key=lambda n: n[1])
print(ordenar_por_precio)
ordenar_por_nom = sorted(productos)
print(ordenar_por_nom)

palabras = ["perro","uva", "router","pagina"]

longitud = sorted(palabras, key=lambda n: len(n))
print(longitud)

numeros = [12, 25, 10, 35, 48]

par_impar = sorted(numeros, key=lambda n: n % 2 != 0)
print(par_impar)


palabra = "algoritmo"

frecuencias = {}
for letra in palabra:
    frecuencias[letra] = frecuencias.get(letra, 0) + 1

ordenado = dict(sorted(frecuencias.items(), key=lambda x: x[1], reverse=True))
print(ordenado)

empleados = [
      ("Maria", "Ventas", 30),
      ("Luis", "Tecnología", 25),
      ("Ana", "Ventas", 25),
      ("Pedro", "Tecnología", 28)
]

ordenar_dep = sorted(empleados, key=lambda n: (n[1], n[2]))
for empleado in ordenar_dep:
    print(empleado)
```
