```python
# Crear una lambda que reciba un número y devuelva True si es múltiplo de 3, False en caso contrario.
multiplo_3 = lambda n: n % 3 == 0
print(multiplo_3(27))
# Crear una lambda que reciba un número y devuelva su cubo.
cubo = lambda n: n ** 3
print(cubo(2))
# Crear una lambda que reciba dos números y devuelva su producto.
producto = lambda x, y: x * y
print(producto(2,3))
# Crear una lambda que reciba dos números y devuelva el mayor.
mayor = lambda x, y: x if x > y else y
print(mayor(5,3))
# Crear una lambda que reciba una palabra y devuelva True si empieza con la letra “A" (o “a”).
primera = lambda x: x.startswith("A") or x.startswith("a")
print(primera("alfredo"))
# Crear una lambda que reciba una temperatura en Celsius y la convierta a Fahrenheit.
cambioT = lambda n: (n * 1.8) + 32
print(cambioT(40))
```
